```julia
using LinearAlgebra
using Plots
```

Compuerta Hadamar es

$$
H=\frac{1}{\sqrt{2}}
\left(\begin{array}{cc} 
1 & 1\\ 
1 & -1
\end{array}\right)
$$

```julia
H=1/sqrt(2)*[1 1; 1 -1] #Definimos la compuerta HAdamar
ϕ=1/sqrt(2)*[1, 1*im] #DEfinimos el cubit inicial

kb=[[1,0],[0,1]] #BASE USUAL
```

```julia
n=3
P=2n+1 #n-1 es el número central

A=Array{Vector{ComplexF64}}(undef, P) #Numero de qbits
B=Array{Vector{ComplexF64}}(undef, P)
```

```julia
@inbounds for i=1:P
    A[i]=[0,0]
    B[i]=[0,0]
end;
```

```julia
A[n+1]=1/sqrt(2)*[1,im] #El elemeno n-1 es el principio "central" [1|0>+i|1>] (X) |0x>
```

```julia
for i=(n+1)-0:(n+1)+0
    if kb[1]*transpose(kb[1])*(H*A[i]) != [0+0im,0+0im]
        B[i-1]=B[i-1]+kb[1]*transpose(kb[1])*(H*A[i])
    end
    if kb[2]*transpose(kb[2])*(H*A[i]) != [0+0im,0+0im]
        B[i+1]=B[i+1]+kb[2]*transpose(kb[2])*(H*A[i])
    end
    for j=(n+1)-0:(n+1)+0
        A[i]=[0,0]
    end
end
```

```julia
n=500
P=2n+1 #n-1 es el número central

A=Array{Vector{ComplexF64}}(undef, P) #Numero de qbits
B=Array{Vector{ComplexF64}}(undef, P)
@inbounds for i=1:P
    A[i]=[0,0]
    B[i]=[0,0]
end;
A[n+1]=1/sqrt(2)*[1,im]
```

```julia
for k=0:n
   
    if k%2!=0
    
        for i=(n+1)-k:(n+1)+k
            if kb[1]*transpose(kb[1])*(H*A[i]) != [0+0im,0+0im]
                B[i-1]=B[i-1]+kb[1]*transpose(kb[1])*(H*A[i])
            end
            if kb[2]*transpose(kb[2])*(H*A[i]) != [0+0im,0+0im]
                B[i+1]=B[i+1]+kb[2]*transpose(kb[2])*(H*A[i])
            end
            for j=(n+1)-k:(n+1)+k
                A[i]=[0,0]
            end
        end
        
    else
        
        for i=(n+1)-k:(n+1)+k
            if kb[1]*transpose(kb[1])*(H*B[i]) != [0+0im,0+0im]
                A[i-1]=A[i-1]+kb[1]*transpose(kb[1])*(H*B[i])
            end
            if kb[2]*transpose(kb[2])*(H*B[i]) != [0+0im,0+0im]
                A[i+1]=A[i+1]+kb[2]*transpose(kb[2])*(H*B[i])
            end
            for j=(n+1)-k:(n+1)+k
                B[i]=[0,0]
            end    
        end
    end

end
```
